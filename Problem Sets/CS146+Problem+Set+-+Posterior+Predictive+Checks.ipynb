```python
from scipy.stats import beta
import random

n_samples = 10000
#out test statistic will be the proportion of successful trials
proportions = []
#here we are sampling from the beta distribution - the parameters were taken from the
#medical trial pre-class work notebook
for i in range(n_samples):
    proportions.append(beta.rvs(58, 65))
 
plt.figure(figsize=(12,7))
plt.hist(proportions, edgecolor='black', color='cornflowerblue', bins=20)
#here we are plotting the test statistic from the real dataset
plt.axvline(x=57/178, label='Proportion of successful trials in the real dataset', color='red')
plt.xlabel('Proportion of successful trials', size=15)
plt.ylabel('Frequency', size=15)
plt.legend()
plt.show()

#calculating the proportion of times test statistic from the replicated datasets is larger
#than the test statistic from the real dataset
print('The p-value is:', round((len([i for i in proportions if i > 57/178]))/n_samples,5))
```
