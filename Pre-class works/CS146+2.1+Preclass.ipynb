```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
x = np.linspace(-8, 4, 100)
plt.plot(x, stats.norm.pdf(x, loc=-2, scale=2), label='loc = -2, scale = 2')
plt.plot(x, stats.norm.pdf(x, loc=-1, scale=0.2), label='loc = -1, scale = 0.2')
plt.legend()
```

```python
dist = stats.norm(loc=-2, scale=2)
data = dist.rvs(size=100)
plt.hist(data, density=True)
plt.plot(x, stats.norm.pdf(x, loc=-2, scale=2), label='loc = -2, scale = 2')
plt.show()
```

```python
x = np.linspace(0, 6, 200)
plt.plot(x, stats.gamma.pdf(x, a=2, scale=1/1), label='a = 2, scale = 1/1')
plt.plot(x, stats.gamma.pdf(x, a=0.1, scale=10), label='a = 0.1, scale = 10/1')
plt.legend()
```

```python
dist = stats.gamma(a=2, scale=1/1)
data = dist.rvs(size=100)

plt.plot(x, stats.gamma.pdf(x, a=2, scale=1/1), label='a = 2, scale = 1/1')
plt.hist(data, density=True)
plt.show()
```
